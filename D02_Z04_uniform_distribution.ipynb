# Rozkład jednostajny na odcinku $[a,b]$
Gęstość zmiennej losowej $X$ o rozkładzie jednostajnym na odcinku $[a,b]$ jest dana przez 
$$
\chi_{[a,b]}(x) = \left\{ \begin{array}{ll}
\frac{1}{b-a} & \textrm{gdy}  x \in [a,b]\\
0 & \textrm{gdy } x \notin [a,b]
\end{array} \right. ,
$$
gdzie $a<b$ oraz $a,b \in \mathbb{R}$.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
import seaborn as sns
```

# Definiujemy zmienną losową

```python
t = np.arange(-5, 5, 0.05)
uniform=stats.uniform(-1,2)
```

# Narysujmy gętość zmiennej losowej

```python
fig, ax = plt.subplots(1, 1)
plt.plot(t, uniform.pdf(t), '--', ms=8, label='frozen pdf')
ax.set_xlim([-5,5])
ax.set_ylim([-0.1,1])
plt.show()
```

# Narysujmy dystrybuantę zmiennej losowej

```python
fig, ax = plt.subplots(1, 1)
ax.plot(t, uniform.cdf(t), '--', ms=8, label='bernoulli cdf')
plt.show()
```

# Wylosujmy próbkę i narysujmy histogram na jednym obrazku

```python
r = uniform.rvs(size=1000)
fig, ax = plt.subplots(1, 1)
ax.hist(r, normed=True, histtype='stepfilled', alpha=0.5, label='histogram')
ax.legend(loc='best', frameon=False)
ax.plot(t, uniform.pdf(t), 'k-', lw=2, label='a=-1, b=1')
ax.set_xlim([-5,5])
ax.set_ylim([-0.1,1])
ax.legend()
plt.show()
```

# Narysujmy gęstości rozkładu jednostajnego z różnymi parametrami

```python
a_values = [1, 1, 1]
b_values = [1, 2, 3]
fig, ax = plt.subplots(1, 1)
for a, b in zip(a_values, b_values):
    dist = stats.uniform(a, b)
    plt.plot(t, dist.pdf(t), ls="--", c='black', label=r'$a=%1.f,\ b=%.1f$' % (a, a+b))
ax.set_xlim([-5,5])
ax.set_ylim([-0.1,1.1])    
plt.legend()
plt.show()
```

# Narysujmy kilka histogramów różnych próbek z rozkładu jednostajnego o tych samych parametrach.

```python
import warnings
warnings.filterwarnings("ignore")

uniform=stats.uniform(0,1)
fig, ax = plt.subplots(nrows=5, ncols=5, figsize=(10, 10))

for i in np.arange(0,5):
    for j in np.arange(0,5):
        r = uniform.rvs(size=100)
        ax[i,j].hist(r, normed=True)
plt.show()
```
