# Rozkład chi kwadrat z $n$ stopniami swobody
Rozkład chi kwadrat ma gęstość:

$$
f_{X}(x) = 
\left\{ \begin{array}{ll}
{\frac {1}{2^{\frac {n}{2}}\Gamma \left({\frac {n}{2}}\right)}}\;x^{{\frac {n}{2}}-1}e^{-{\frac {x}{2}}} 
   & \textrm{gdy } x \geq 0\\
0 & \textrm{gdy } x < 0 
\end{array} \right. ,
$$

gdzie $\Gamma$ oznacza funkcję Gamma https://pl.wikipedia.org/wiki/Funkcja_\%CE\%93


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
import seaborn as sns
```

# Definiujemy zmienną losową

```python
t = np.arange(0, 20, 0.05)
n=4
X=stats.chi2(n)
```

# Narysujmy gęstość zmiennej losowej

```python
#gestosc
# zbior, pdf od zmiennej w zakresie t
plt.plot(t,X.pdf(t), 'b--')

plt.show()
```

# Narysujmy dystrybuantę zmiennej losowej

```python
#dydtrybucja
plt.plot(t,X.cdf(t), 'b--')

plt.show()
```

# Wylosujmy próbkę i narysujmy histogram na jednym obrazku

```python
import warnings
warnings.filterwarnings("ignore")

data = X.rvs(1000) # ramdomowe losowanie
plt.hist(data, normed=True, histtype='stepfilled', alpha=0.5, label='histogram') # histogram b. wazne 'normed' bo normalizuje
plt.plot(t,X.pdf(t)) # gestosc
plt.legend()
plt.show()
```

# Narysujmy gęstości rozkładu z różnymi parametrami

# Narysujmy kilka histogramów różnych próbek z rozkładu o tych samych parametrach.

```python
import warnings
warnings.filterwarnings("ignore")


fig, ax = plt.subplots(nrows=5, ncols=5, figsize=(10, 10))

for i in np.arange(0,5):
    for j in np.arange(0,5):
        r = X.rvs(size=100)
        ax[i,j].hist(r, normed=True)
plt.show()
```
