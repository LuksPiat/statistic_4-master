# Rozkład normalny
Gęstość zmiennej losowej $X$ o rozkładzie normalnym z parametrami $\mu$ i $\sigma$ jest dana przez 
$$
f(x) = \frac{1}{\sigma \sqrt{2 \pi} } e^{-\frac{(x-m)^2}{2\sigma^2} }.
$$

# Zad.

Definiujemy zmienną losową

```python
from scipy.stats import norm
```

```python
normal = norm(0,1)
```

# Zad.


Narysujmy gętość zmiennej losowej

```python
t=  np.arange(-5,5,0.01)
```

```python
fig, ax = plt.subplots(1, 1)
plt.plot(t, normal.pdf(t), '--', ms=8, label='normal')
ax.set_xlim([-5,5])
ax.set_ylim([-0.1,1])
plt.show()
```

# Zad.


Narysujmy dystrybuantę zmiennej losowej

```python
fig, ax = plt.subplots(1, 1)
ax.plot(t, normal.cdf(t), '--', ms=8, label='bernoulli cdf')
plt.show()
```

# Zad.


Wylosujmy próbkę i narysujmy histogram na jednym obrazku

```python
r = normal.rvs(size=1000)
fig, ax = plt.subplots(1, 1)
ax.hist(r, normed=True, histtype='stepfilled', alpha=0.5, label='histogram')
ax.legend(loc='best', frameon=False)
ax.plot(t, normal.pdf(t), 'k-', lw=2, label='normal distribution')
ax.set_xlim([-5,5])
ax.set_ylim([-0.1,1])
ax.legend()
plt.show()
```

# Zad.


Narysujmy kilka histogramów różnych próbek z roskładu normalnego o tych samych parametrach.

```python
import warnings
warnings.filterwarnings("ignore")

normal=stats.norm(0,1)
fig, ax = plt.subplots(nrows=5, ncols=5, figsize=(10, 10))

for i in np.arange(0,5):
    for j in np.arange(0,5):
        r = normal.rvs(size=100)
        ax[i,j].hist(r, normed=True)
plt.show()
```
