# Inverse Survival Function (ISF):

Pamiętamy, że Survival Function miała postać 

$$
SF(x)=1-CDF(x).
$$

Pamiętamy również, żę Percentile Point Function (PPF) odwrotna do dystrybunanty miała postać: 

$$
PPF(x)=CDF^{-1}(x).
$$

Funkcja PPF pozwalała nam liczyć kwantyle, czyli rozwiązać problem:

Jak znaleźć najmniejszą taką wartość, że jest ona większa od $p$ (np. $p=0.75$) innych liczb. 

Funkcja ISF to funkcja odwrotna do $SF$:

$$
ISF(x)=SF^{-1}(x).
$$

Za pomocą ISF możemy rozwiązać problem:

Jak znaleźć największą taką wartość, że jest ona mniejsza od $p$ (np. $p=0.75$) innych liczb. 


```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
from scipy import integrate
```

Będziemy liczyć kwantyl rzędu $p=0.5$

```python
p=0.75
```

Zdefiniujmy zmienną losową o rozkładzie Normalnym

```python
(mu, sigma)=(0,1)
nd=stats.norm(mu, sigma)
```

# Zadanie 1
Narysuj na oddzielnych wykresach: 
<ul>
<li> gęstość rozkładu normalnego </li>
<li> jeden minus dystrybuanta rozkładu normalnego ($1-CDF(x)$) </li>
<li> funkcję $ISF$ </li>
</ul>

i zaznacz na nich odpowiednie wartości tak, by móc odtworzyć poniższy rysunek.

```python
x = np.linspace(-3,3,100)
yp = nd.pdf(x)
x1 = np.linspace(stats.norm.ppf(1-p),3)
y1 = nd.pdf(x1)
yp2 = 1-stats.norm.cdf(x)
yp3 = stats.norm.isf(x, loc=mu, scale=sigma)




figs, axs = plt.subplots(1,3, figsize=(10, 4))
#gestosc
axs[0].plot(x,yp, 'k')
axs[0].text(stats.norm.ppf(1-p), 0.1, 0.25)
axs[0].fill_between(x1, y1, facecolor='#CCCCCC')
axs[0].hlines(0.33, -3, 3, linestyles='--')
# dystrybuanta
axs[1].plot(x,yp2, 'k')
axs[1].text(stats.norm.cdf(1-p), p,p)
axs[1].hlines(p, -3, 3, linestyles='--')
# funkcja odwrotna od dystrybuanty
axs[2].plot(x,yp3, 'k')
axs[2].text(p, stats.norm.isf(p), np.round(stats.norm.ppf(p), 4))
axs[2].vlines(p, y1, np.round(stats.norm.isf(p),4), linestyles='--')


plt.show()
```

```python
x = np.linspace(-3, 3, 500)
yp = nd.pdf(x)
x1 = np.linspace(stats.norm.ppf(1-p), 3)
y1 = nd.pdf(x1)
yp2 = stats.norm.sf(x)
yp3 = stats.norm.isf(x, loc=mu, scale=sigma)

figs,axs = plt.subplots(1,3, figsize=(12, 6))

axs[0].plot(x, yp, 'k')
axs[0].text(stats.norm.isf(p), 0.1, p)
axs[0].fill_between(x1, y1, facecolor='#CCCCCC')
axs[0].set_xlabel('x')
axs[0].set_ylabel('PDF(x)')

axs[1].plot(x, yp2, 'k')
axs[1].text(stats.norm.sf(p), p, p)
axs[1].hlines(p, -3, 3, linestyles='--')
axs[1].set_xlabel('x')
axs[1].set_ylabel('1-CDF(x)=SF(x)')

axs[2].plot(x, yp3, 'k')
axs[2].text(p, stats.norm.isf(p), np.round(stats.norm.ppf(p), 4))
axs[2].vlines(p, y1, np.round(stats.norm.isf(p),4), linestyles='--')
axs[2].set_xlabel('x')
axs[2].set_ylabel('ISF(x)')

plt.show()
```
